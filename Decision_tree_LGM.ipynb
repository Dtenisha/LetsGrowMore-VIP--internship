```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/ML_datasets/Iris (1).csv')
```

```python
data.head()
```

```python
def data_exploration(data):
  print(data.head())
  print('-----------------------------------------------------------------------------')
  print(data.tail())
  print('-----------------------------------------------------------------------------')
  print(data.shape)
  print('-----------------------------------------------------------------------------')
  print(data.describe())
  print('-----------------------------------------------------------------------------')
  return data.info()
```

```python
data_exploration(data)
```

```python
data['Species'].value_counts()
```

```python
sns.countplot(data['Species'],saturation = 0.75,palette = 'inferno_r')
plt.title('Count plot for Iris Species')
```

```python
sns.pairplot(data.drop(['Id'],axis=1),hue = 'Species',palette='inferno_r')
plt.show()
```

```python
X = data.iloc[:,1:5]
y = data.iloc[:,-1]
```

```python
print(X.head())
print('-----------------------------------------------------------')
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 1/3,random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='gini',min_samples_split=2)
```

```python
dtc
```

```python
print(X.shape)
print(y.shape)
```

```python
dtc=dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_test, y_pred))
```

```python
feature_names = data.columns
feature_names
```

```python
target = (['Iris-versicolor','Iris-virginica','Iris-setosa'])

```

```python
from sklearn import tree
plt.figure(figsize=(20,15))
tree.plot_tree(dtc,filled = True,feature_names=feature_names,rounded=True,class_names=target)
plt.title('Decision Tree for IRIS Dataset')
plt.show()
```
